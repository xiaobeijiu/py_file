# 按掩膜提取

```python
import os
import arcpy
from arcpy import env
from arcpy.sa import *
#允许覆盖已有文件
arcpy.env.overwriteOutput = True

# 设置工作环境
env.workspace = r"D:\新建文件夹\遥感\fvc数据"


# 输出目录
output_folder = r"D:\新建文件夹\遥感\FVC(掩膜)"

# 检查输出目录是否存在，如果不存在则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 获取所有栅格文件
raster_list = arcpy.ListRasters()  # 列出所有栅格文件

mask = r"D:\新建文件夹\遥感\ROI\Tarim.shp"  # 这里假设掩膜是矢量文件

# 遍历栅格文件和掩膜文件
for raster in raster_list:
    
    # 创建掩膜提取
    outExtractByMask = ExtractByMask(raster, mask)
        
    # 定义输出路径
    output_name = f"{raster}" #输出文件名根据需求修改
    output_path = os.path.join(output_folder, output_name)
        
    # 保存结果
    outExtractByMask.save(output_path)
    print(f"输出文件已保存: {output_path}")
```

# 重采样

```python
import arcpy
import os

# 设置工作空间路径
arcpy.env.workspace = r"D:\新建文件夹\遥感图像处理作业\fvc数据"

# 获取参考 TIFF 文件(进行重采样的目标像元大小)
reference_tif = "Temperature_2000.tif"  # 可以修改成目标参考文件名

# 获取参考 TIFF 文件的像素大小
desc = arcpy.Describe(reference_tif)

# 提取像素宽度和高度
pixel_size = desc.meanCellWidth  # 获取像素的宽度

# 获取所有的 .tif 文件
tif_files = arcpy.ListFiles("*.tif")

# 设置输出路径
output_path = r"D:\新建文件夹\遥感图像处理作业\fvc重采样"

# 循环处理每个 .tif 文件
for tif in tif_files:
    if tif != "Temperature_2000.tif":  # 跳过参考文件（将参考文件与源文件放在一个路径下）
        # 构建输出文件的路径
        output_file = os.path.join(output_path, "resampled_" + tif)
        
        # 执行重采样，使用参考文件的像素大小
        arcpy.Resample_management(tif, output_file, str(pixel_size), "NEAREST")#四个参数
        #1. 输入数据
        #2. 输出数据
        #3. 像素大小（字符串格式）
        #4. 重采样方法（NEAREST表示最近邻重采样） BILINEAR表示双线性重采样，CUBIC表示立方卷积重采样 MAJORITY表示多数重采样
        
        print(f"已重采样: {tif} 到 {output_file}")

```

# 条件函数

```python
# 描述: 对输入目录下所有栅格文件的每个像元执行条件 if/else 评估
# 要求: 需要 Spatial Analyst 扩展模块

import arcpy
from arcpy import env
from arcpy.sa import *
import os

# 设置工作空间（栅格文件所在目录）
env.workspace = r"D:\新建文件夹\遥感\土壤湿度（已处理）"  # 替换为你的路径

# 设置输出目录
output_dir = r"D:\新建文件夹\遥感\土壤湿度（已已处理）" # 替换为你的输出路径
# 确保输出目录存在，如果不存在则创建
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 确保Spatial Analyst扩展模块可用
arcpy.CheckOutExtension("Spatial")

# 获取目录中的所有栅格文件
raster_list = arcpy.ListRasters("*.tif")  # 假设栅格为tif格式，根据需要调整

# 允许覆盖输出
arcpy.env.overwriteOutput = True

# 批量处理所有栅格文件
for raster in raster_list:
    # 拼接完整路径
    in_raster_path = os.path.join(env.workspace, raster)
    # 读取栅格
    inRaster = Raster(in_raster_path)

    # 设置条件表达式
    # 这里可以用 where 子句或者直接在条件中使用表达式
    # 例如：VALUE>=1500
    whereClause = "VALUE >= 0"

    # 使用Con函数执行条件判断：满足条件的像元赋值为1，否则为0
    outCon = Con(inRaster, inRaster, 0, whereClause)

    # 输出文件路径
    output_path = os.path.join(output_dir, f"{raster}")

    # 保存结果
    outCon.save(output_path)
    print(f"已处理: {raster} 到 {output_path}")
# 释放扩展
arcpy.CheckInExtension("Spatial")
```

# 裁剪

```python
import arcpy
import os

# 设置工作空间（shp文件所在文件夹）
arcpy.env.workspace = r"D:\新建文件夹\TOD\poi数据\分类（合并）"

# 设置裁剪范围shp
clip_features = r"D:\新建文件夹\TOD\TOD(gis)\TOD\TOD.gdb\四环边界线"

# 设置输出文件夹
output_folder = r"D:\新建文件夹\TOD\poi数据\裁剪结果"
#确保输出目录存在，如果不存在则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 获取所有shp文件
shp_list = arcpy.ListFeatureClasses("*.shp")

# 循环裁剪
for shp in shp_list:
    # 输出文件路径
    out_shp = os.path.join(output_folder, f"clip_{shp}")
    # 执行裁剪
    arcpy.analysis.PairwiseClip(shp, clip_features, out_shp)#被裁剪的、裁剪范围，输出文件
    print(f"已裁剪: {shp} -> {out_shp}")# 导入系统模块
```

# 路网裁剪

```python
import arcpy
import os

# 允许覆盖输出文件
arcpy.env.overwriteOutput = True

# 设置工作空间（shp文件所在文件夹）
arcpy.env.workspace = r"D:\新建文件夹\2000"

# 设置裁剪要素（道路数据）
input_shp = r"D:\新建文件夹\2000\gis_osm_roads_free_1.shp"

# 设置裁剪范围shp（这里是县.shp）
clip_features = r"D:\1文件\arcgis\2024年省市县三级行政区划数据（审图号：GS（2024）0650号）\shp格式的数据（调整过行政区划代码，补全省市县信息）\县.shp"

# 设置输出文件夹
output_folder = r"D:\新建文件夹\路网数据"

# 确保输出目录存在，如果不存在则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 定义属性查询条件，选择符合条件的县
query = "市名 = '张掖市'"  # 替换为你要选择的字段和条件

# 提取文件名（不包含路径）
clip_name = os.path.basename(clip_features)

# 创建临时的选择文件
selected_clip_shp = os.path.join(output_folder, "selected_counties.shp")

# 使用 Select_analysis 来根据属性查询选择县.shp 中符合条件的要素
arcpy.Select_analysis(clip_features, selected_clip_shp, query)

# 生成输出文件路径（裁剪后的道路数据）
out_shp = os.path.join(output_folder, "2000_张掖市.shp")

# 执行裁剪操作，使用选择后的县作为裁剪范围
arcpy.analysis.Clip(input_shp, selected_clip_shp, out_shp)

# 删除中间步骤中的临时文件，只保留最终的裁剪结果
arcpy.Delete_management(selected_clip_shp)

# 输出裁剪结果
print(f"已裁剪: {input_shp} -> {out_shp}")
```

# 修改属性表里的字段名

```python
import arcpy
import os

# 设置工作空间，包含要素类的文件夹路径
workspace = r"D:\新建文件夹\TOD\缓冲区（已处理）\poi.gdb"
arcpy.env.workspace = workspace

# 获取工作空间中的所有要素类
feature_classes = arcpy.ListFeatureClasses()

# 定义需要修改的字段名称
old_field_name = "个数"  # 旧字段名

# 遍历每个要素类
for fc in feature_classes:
    try:
        # 提取图层名称，用作新字段名
        layer_name = os.path.splitext(fc)[0]
        # 检查要素类是否包含目标字段
        if arcpy.ListFields(fc, old_field_name):
            # 修改字段名
            arcpy.AlterField_management(fc, old_field_name, layer_name)
            print(f"字段 {old_field_name} 在 {fc} 中已修改为 {layer_name}")
        else:
            print(f"字段 {old_field_name} 在 {fc} 中不存在")
    except Exception as e:
        print(f"处理 {fc} 时出错: {e}")
```

# 以表格显示分区统计

```python
import arcpy
import os

# 设置工作环境
arcpy.env.workspace = r"D:\新建文件夹\2000"  # 数据文件夹路径
output_folder = r"D:\新建文件夹\2000统计有"  # 输出表格文件夹路径

# 允许覆盖输出文件
arcpy.env.overwriteOutput = True

# 确保输出目录存在，如果不存在则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 获取所有的栅格数据和分区数据
raster_list = arcpy.ListRasters("*.tif", "TIF")  # 根据需要设置文件格式和名称模式
zone_shapefile = r"D:\新建文件夹\渔网投影\fishnet_lin.shp"  # 分区面数据路径

# 遍历每个栅格数据，进行分区统计
for raster in raster_list:
    # 输出表格路径
    output_table = os.path.join(output_folder, f"{os.path.splitext(raster)[0]}.dbf")
    
    # 执行分区统计
    arcpy.sa.ZonalStatisticsAsTable(zone_shapefile,  # 分区面数据
                                    "grid_id",      # 分区字段
                                    raster,          # 输入栅格数据
                                    output_table,    # 输出表格路径
                                    "NODATA",        # 处理无数据值
                                    "MEAN")           # 统计类型（ALL表示所有统计量）
       
    print(f"完成栅格：{raster}，输出表格：{output_table}")

print("所有分区统计完成！")
                                                                                                    
```

# xy转点

```python
import arcpy
import os

# 设置工作空间
workspace = r"D:\新建文件夹\合并后的热力图数据"  # CSV文件所在的文件夹路径
output_folder = r"D:\新建文件夹\热力图数据"  # 输出的Shapefile文件夹路径

# 允许覆盖输出文件
arcpy.env.overwriteOutput = True

# 检查输出文件夹是否存在，如果不存在则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 获取工作空间中所有以 .csv 结尾的文件
csv_files = [f for f in os.listdir(workspace) if f.endswith('.csv')]

# 遍历每个CSV文件
for csv_file in csv_files:
    # 构建CSV文件的完整路径
    csv_path = os.path.join(workspace, csv_file)
    
    # 构建输出Shapefile的路径
    output_shapefile = os.path.join(output_folder, f"{os.path.splitext(csv_file)[0]}.shp")

    # 使用XYTableToPoint工具将CSV文件转换为点要素
    # 参数说明：
    # 1. csv_path: 输入的CSV文件路径
    # 2. output_shapefile: 输出的Shapefile路径
    # 3. "X": 表示CSV文件中存储X坐标的字段名
    # 4. "Y": 表示CSV文件中存储Y坐标的字段名
    # 5. coordinate_system: 设置坐标系统为WGS 1984（可以根据需要修改）
    arcpy.management.XYTableToPoint(csv_path, 
                                    output_shapefile, 
                                    "wgs84_LNG", 
                                    "wgs84_LAT",
                                    coordinate_system="GCS_WGS_1984")  # 设置坐标系统为WGS 1984

    # 打印转换成功的信息
    print(f"成功转换：{csv_file} -> {output_shapefile}")

# 打印所有文件处理完成的信息
print("所有文件处理完成！")
```

# 核密度分析

```python
import arcpy
from arcpy import env
from arcpy.sa import *
import os

# 设置工作空间
env.workspace = r"D:\新建文件夹\TOD\poi数据\裁剪结果"  # shp文件所在的文件夹路径
output_folder = r"D:\新建文件夹\TOD\POI核密度"  # 输出的Shapefile文件夹路径

# 允许覆盖输出文件
arcpy.env.overwriteOutput = True

# 检查输出文件夹是否存在，如果不存在则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 获取所有shp文件
shapefiles = arcpy.ListFeatureClasses("*.shp")
# 遍历每个Shapefile进行核密度分析
for shapefile in shapefiles:
    # 构建Shapefile的完整路径
    input_shapefile = os.path.join(env.workspace, shapefile)
    # 设置输出路径和文件名
    output_raster = os.path.join(output_folder, f"{shapefile[:-4]}.tif")

    # 执行核密度分析
    out_kernel_density = KernelDensity(input_shapefile,  # 输入要素
                                       "NONE",          # 计算密度的字段（如果没有字段，设置为"NONE"）
                                       cell_size=0.0008,     # 输出栅格的像素大小（单位为米）
                                       area_unit_scale_factor="SQUARE_KILOMETERS",  # 区域单位缩放因子（单位为平方千米）
                                       out_cell_values="DENSITIES",  # 输出单元格值类型（密度）
                                       method="PLANAR",   # 方法（平面或地理）
                                       in_barriers=None        # 输入障碍物（如果没有障碍物，设置为None）
                                        ) # 搜索半径（单位为米）

    # 保存结果到指定路径
    out_kernel_density.save(output_raster)

    print(f"成功处理：{shapefile} -> {output_raster}")

print("所有核密度分析完成！")
    
```

# 合并

```python
import arcpy
import os
# 允许覆盖输出
arcpy.env.overwriteOutput = True
# 设置工作空间
arcpy.env.workspace = r"D:\新建文件夹\poi数据\教育机构"  # 替换为实际数据所在路径

# 获取需要合并的要素类列表
feature_classes = arcpy.ListFeatureClasses()

# 检查是否有要素类可以合并
if not feature_classes:
    print("未找到要素类")
else:
    # 设置输出路径和名称
    output = r"D:\新建文件夹\poi数据\分类（合并）\教育机构.shp"  # 替换为期望的输出路径
    # 检查输出路径是否存在，如果不存在则创建
    if not os.path.exists(os.path.dirname(output)):
        os.makedirs(os.path.dirname(output))

    
    # 执行合并操作
    arcpy.management.Merge(
        inputs=feature_classes,      # 输入要素类列表
        output=output,               # 输出要素类路径
        field_mappings=None,           # 字段映射，空字符串表示使用默认映射
        add_source="NO_SOURCE_INFO" , # 是否添加源数据信息，NO_SOURCE_INFO表示不添加
        field_match_mode="AUTOMATIC"  # 字段匹配模式，AUTOMATIC表示自动匹配
    )
    
    print(f"合并完成，输出文件：{output}")
    print(f"共合并 {len(feature_classes)} 个要素类")


```
