```python
import pandas as pd
import numpy as np

# 读取上传的Excel文件
file_path = r"D:\新建文件夹\Q4.csv" 
df = pd.read_csv(file_path,encoding='gbk')
feature_columns = ['MD19_34', 'MD35_54', 'MD55_64', 'sex', 'LW', 'SHDI', 'D2Tran', 'Per_Dis', 'ZZ']
data = df[feature_columns]
# Z-Score 标准化函数
def z_score_normalization(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    normalized_data = (data - mean) / std_dev
    return normalized_data

# 对所有列进行Z-Score标准化
for column in data.columns:
    data[column] = z_score_normalization(data[column])

# 查看标准化后的数据
print(data.head())

# 保存标准化后的数据到新的CSV文件
output_file_path = r"D:\新建文件夹\z-score标准化.csv" 
data.to_csv(output_file_path, index=False)

# 输出保存的文件路径
output_file_path
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
df=pd.read_csv(r"D:\新建文件夹\z-score标准化.csv")

# 绘制数据分布直方图
plt.figure(figsize=(10, 6))
plt.hist(df, bins=30, density=True, alpha=0.7)
plt.xlabel('数值')
plt.ylabel('频率')
plt.title('数据分布直方图')


# 计算基本统计量
print("数据基本统计信息：")
print(f"均值: {np.mean(df):.2f}")
print(f"标准差: {np.std(df):.2f}")
print(f"中位数: {np.median(df):.2f}")   
print(f"偏度: {stats.skew(df):.2f}")
print(f"峰度: {stats.kurtosis(df):.2f}")
```
